```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr
```

```python
df = pd.read_excel('PesquisaAuto.xlsx', sheet_name=0, usecols="A:K")
```

# Análise Exploratória - Centro e dispersão

## Variável IMAGEM

```python

print("Média: \t\t\t\t| ",df["Imagem"].describe()['mean'])
print("Desvio Padrão: \t\t\t| ",df["Imagem"].describe()['std'])
print("Valor mínimo: \t\t\t| ",df["Imagem"].describe()['min'])
print("25% : \t\t\t\t| ",df["Imagem"].describe()['25%'])
print("Mediana: \t\t\t| ",df["Imagem"].describe()['50%'])
print("75%: \t\t\t\t| ",df["Imagem"].describe()['75%'])
print("Valor Máximo: \t\t\t| ",df["Imagem"].describe()['max'])
print("Amplitude: \t\t\t| ", df["Imagem"].max() - df["Imagem"].min())
print("Intervalo Interquartil: \t| ", df["Imagem"].describe()['75%'] - df["Imagem"].describe()['25%'])
print("Cerca 1: \t\t\t| ", df["Imagem"].describe()['25%'] - (1.5*(df["Imagem"].describe()['75%'] - df["Imagem"].describe()['25%'])))
print("Cerca 2: \t\t\t| ", df["Imagem"].describe()['75%'] + (1.5*(df["Imagem"].describe()['75%'] - df["Imagem"].describe()['25%'])))
print("Coeficiente de variação: \t| ", df["Imagem"].describe()['std']/df["Imagem"].describe()['mean'])

```

## Variável PREÇO

```python

print("Média: \t\t\t\t| ",df["Preco"].describe()['mean'])
print("Desvio Padrão: \t\t\t| ",df["Preco"].describe()['std'])
print("Valor mínimo: \t\t\t| ",df["Preco"].describe()['min'])
print("25% : \t\t\t\t| ",df["Preco"].describe()['25%'])
print("Mediana: \t\t\t| ",df["Preco"].describe()['50%'])
print("75%: \t\t\t\t| ",df["Preco"].describe()['75%'])
print("Valor Máximo: \t\t\t| ",df["Preco"].describe()['max'])
print("Amplitude: \t\t\t| ", df["Preco"].max() - df["Preco"].min())
print("Intervalo Interquartil: \t| ", df["Preco"].describe()['75%'] - df["Preco"].describe()['25%'])
print("Cerca 1: \t\t\t| ", df["Preco"].describe()['25%'] - (1.5*(df["Preco"].describe()['75%'] - df["Preco"].describe()['25%'])))
print("Cerca 2: \t\t\t| ", df["Preco"].describe()['75%'] + (1.5*(df["Preco"].describe()['75%'] - df["Preco"].describe()['25%'])))
print("Coeficiente de variação: \t| ", df["Preco"].describe()['std']/df["Preco"].describe()['mean'])
```

## Variável UTILITÁRIO

```python

print("Média: \t\t\t\t| ",df["Utilitario"].describe()['mean'])
print("Desvio Padrão: \t\t\t| ",df["Utilitario"].describe()['std'])
print("Valor mínimo: \t\t\t| ",df["Utilitario"].describe()['min'])
print("25% : \t\t\t\t| ",df["Utilitario"].describe()['25%'])
print("Mediana: \t\t\t| ",df["Utilitario"].describe()['50%'])
print("75%: \t\t\t\t| ",df["Utilitario"].describe()['75%'])
print("Valor Máximo: \t\t\t| ",df["Utilitario"].describe()['max'])
print("Amplitude: \t\t\t| ", df["Utilitario"].max() - df["Utilitario"].min())
print("Intervalo Interquartil: \t| ", df["Utilitario"].describe()['75%'] - df["Utilitario"].describe()['25%'])
print("Cerca 1: \t\t\t| ", df["Utilitario"].describe()['25%'] - (1.5*(df["Utilitario"].describe()['75%'] - df["Utilitario"].describe()['25%'])))
print("Cerca 2: \t\t\t| ", df["Utilitario"].describe()['75%'] + (1.5*(df["Utilitario"].describe()['75%'] - df["Utilitario"].describe()['25%'])))
print("Coeficiente de variação: \t| ", df["Utilitario"].describe()['std']/df["Utilitario"].describe()['mean'])
```

## Boxplot

## Scatterplot

```python
quanti = df[["Imagem","Preco","Utilitario", "Prototipo"]]
sns.pairplot(df, vars=["Imagem", "Preco", "Utilitario"], kind="scatter", hue="Prototipo", palette="Set2")
plt.show()
```

## Histograma

```python
f, axes = plt.subplots(2,2, figsize=(7,7))
sns.distplot(df['Imagem'], color="blue", ax=axes[0,0], bins=15)
sns.distplot(df['Preco'], color="red", ax=axes[0,1], bins=15)
sns.distplot(df['Utilitario'], color="green", ax=axes[1,0], bins=15)
plt.show()
```

## Distribuição de Frequências

```python
ax = sns.barplot(x="Prototipo", y="Prototipo", data=df, estimator=lambda x: len(x) / len(df) * 100)
ax.set(ylabel="Percentual")

plt.show()
```

## Variáveis vs Imagem
